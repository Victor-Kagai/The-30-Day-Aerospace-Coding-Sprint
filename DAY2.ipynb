We'll try to track threats in a given 3D space continuum.The positions will be more of gps locations so that's why we take it as single point in space not something you append to.

```python
# COMBAT-READY THREAT TRACKING:
threat_tracker = {
    "orion": {
        "current_position": (37.7749, -122.4194, 500.0),  # Tuple - immutable
        "position_history": [(37.7749, -122.4194, 500.0)],  # List of tuples
        "altitude": 500.0,
        "velocity": (250.0, 45.0, 0.0),  # 3D velocity vector
        "last_update": 123456789.0
    },
    "loki": {
        "current_position": (120.123, 76.456, 430.0),
        "position_history": [(120.123, 76.456, 430.0)],
        "altitude": 430.0,
        "velocity": (180.0, -30.0, 25.0),
        "last_update": 123456790.0
    }
}

# CORRECT WAY TO UPDATE POSITION HISTORY:
def update_threat_position(threat_id, new_position):
    threat = threat_tracker[threat_id]
    threat["position_history"].append(new_position)  # Add to history list
    threat["current_position"] = new_position  # Update current position
    
    # Keep only last 100 positions (memory management)
    if len(threat["position_history"]) > 100:
        threat["position_history"].pop(0)

# USAGE:
update_threat_position("orion", (37.7750, -122.4195, 495.0))
update_threat_position("loki", (120.124, 76.455, 425.0))
```
